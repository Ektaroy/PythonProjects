# **Visualization Analysis**

```python
### Histogram (Frequency Distribution)

plt.title("Histogram")
df['Product_line'].hist(bins=20)
plt.xticks(rotation=90,fontsize=10)
plt.figure(figsize=(25,5))
plt.tight_layout()
plt.show()
```

```python
### Box Plot (Detect Outliers)

plt.title("Box Plot")
sns.boxplot(x=df['Rating'])
plt.show()
```

```python
### Barplot

plt.figure(figsize=(8,6))
plt.title("Bar plot")
sns.barplot(x=df['Total'],y=df['Payment'])
plt.show()
```

```python
### countplot

sns.countplot(x=df['Branch'],hue=df['Rating'])
plt.title("countplot")
plt.show()
```

```python
### scatter plot

plt.scatter(df.Quantity,df.Unit_price,s=10)
plt.title("scatter plot")
plt.colorbar()
plt.show()
```

```python
### Rel plot

sns.relplot(x=df['City'],y=df['Rating'],s=10)
plt.title("Rel plot")
plt.show()
```

```python
### Jointplot (Scatter Plot + Histogram)

sns.jointplot(x=df['Product_line'],y=df['gross_income'],data=df,s=50,color='green')
plt.title("Jointplot")
plt.show()
```

```python
sns.jointplot(x=df['Total'],y='Payment',data=df,s=20,color='red')
plt.title("Jointplot")
plt.show()
```

```python
### bar graph

px.bar(df,x='Product_line',y='gross_margin_percentage')
```

```python
### pie chart

plt.title("pie chart")
df['Gender'].value_counts().plot(kind='pie', figsize=(6,6))
```

```python
df['Customer_type'].value_counts().plot(kind='pie', figsize=(6,6))
plt.title("pie chart")
```

```python
### pairplot

sns.pairplot(df,hue='gross_income')
plt.show()
```

```python
### PairGrid

a = sns.PairGrid(df)
a.map_diag(sns.histplot)
a.map_offdiag(sns.scatterplot)
plt.show()
```
